```python
# Boxplots for Visual Outlier Detection

# List of numeric columns to check for outliers
numeric_cols = ["credit_score", "age", "tenure", "balance", "estimated_salary", "products_number"]

# Plot boxplots
plt.figure(figsize=(16, 12))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(3, 2, i)
    sns.boxplot(y=df[col])
    plt.title(f"Boxplot of {col}")
plt.tight_layout()
plt.show()
```

```python
numeric_cols = ['age', 'balance', 'estimated_salary', 'credit_score']
def remove_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]
```

```python
df_cleaned = df.copy()
for col in numeric_cols:
    df_cleaned = remove_outliers_iqr(df_cleaned, col)
print("Original shape:", df.shape)
print("Cleaned shape after outlier removal:", df_cleaned.shape)
```

```python
# Visualize Before vs After

for col in numeric_cols:
    plt.figure(figsize=(12, 4))

    # Before removal
    plt.subplot(1, 2, 1)
    sns.boxplot(y=df[col])
    plt.title(f"{col} - Before")

    # After removal
    plt.subplot(1, 2, 2)
    sns.boxplot(y=df_cleaned[col])
    plt.title(f"{col} - After")

    plt.tight_layout()
    plt.show()
```

```python
# Count of Removed Records Per Column after cleaning Outliers

for col in numeric_cols:
    original_len = len(df)
    cleaned_len = len(remove_outliers_iqr(df, col))
    print(f"{col}: {original_len - cleaned_len} outliers removed")
```

```python
#Saving changes to new CSV file

df_cleaned.to_csv("bank_churn_final.csv", index=False)

from google.colab import files
files.download("bank_churn_final.csv")
```
